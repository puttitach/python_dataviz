### ตัดคอลัมสุดท้าย อยากได่ตัวเลขอย่างเดียว จะตัดโดยใช้คล้าย list [แถว , หลัก]

```python
data_np = irisdata.iloc[ : , :-1].to_numpy()
data_np
```

## ตัดคลาส ตั้งแต่ตัวแรกถึงตัวที่ห้าสิบ

```python
setosa_np = irisdata.iloc[ : 50 , :-1].to_numpy()
setosa_np
```

```python
versicolor_np = irisdata.iloc[ 50 : 100 , :-1].to_numpy()
versicolor_np
virginica_np = irisdata.iloc[ 50 : 100 , :-1].to_numpy()
virginica_np 
```

### visualize array

```python
from matplotlib import pyplot as plt
import matplotlib
```

### สีเข้มแทนค่าน้อย สีอ่อนแทนค่ามาก
### ทำการแปลง array ให้เป็นรูป

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.imshow(setosa_np, interpolation= 'nearest')
plt.show()
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.imshow(versicolor_np, interpolation= 'nearest')
plt.show()
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.imshow(virginica_np, interpolation= 'nearest')
plt.show()
```

#### plot หลายๆรูปไว้ในรูปเดียว (1, 3 ,1) -> plot 1 แถว, 3 หลัก, ตำแหน่งที่ 1

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.subplot(1, 3, 1)
plt.imshow(setosa_np, interpolation= 'nearest')
plt.subplot(1, 3, 2)
plt.imshow(versicolor_np, interpolation= 'nearest')
plt.subplot(1, 3, 3)
plt.imshow(virginica_np, interpolation= 'nearest')
plt.show()
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.imshow(data_np, interpolation= 'nearest')
plt.show()
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.imshow(setosa_np, interpolation= 'nearest', cmap=plt.get_cmap('GnBu'))
plt.show()
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [12, 12 ]
plt.imshow(setosa_np, interpolation= 'nearest', cmap=plt.get_cmap('Pastel2'))
plt.show()
```
