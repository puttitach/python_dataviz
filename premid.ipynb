ข้อ1 สร้าง function ชื่อ createDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
- ถ้า True จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า False จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

เช่น

output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )

จะได้ output1 เป็น { 'apple' : 1 , 'banana' : 2 }

output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )

จะได้ output2 เป็น { 1 : 'apple' , 2 : 'banana' }

### 1 สร้างฟังก์ชั่นก่อน ชื่อcreateDict 
### function คือ lish

```python
# วิธีอาจารย์
def createDict(listtext, listnumber, logicV) : 
    if logicV:  #logicV == True
        retdict = {}
        for index in range(len(listtext)) :
            retdict[listtext[index]] = listnumber[index]   #index คือ
    else :
        retdict = {}
        for index in range(len(listtext)) :
            retdict[listnumber[index]] = listtext[index]
    return retdict        
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
```

```python
# วิธีเบิร์ด
def createDict(k1,k2,x):
    f = {}
    for i in range(len(k1)):
        if x==True:
            f[k1[i]] = k2[i]
        else:
            f[k2[i]] = k1[i]
    return(f)
```

```python
createDict(['apple','banana'],[1,2],True)
```

#### ข้อ2 how many time A, T, G, C in the sequence? (บอกว่าแต่ละกรดอะมิโนในแต่ละ A, T, G, C มีกี่ตัว)

```python
def countChar(seq, character) :   #seq คือสาย ให้เก็บอะไรเข้าไป , char รับว่าเราจะดูตัวอะไร
    charcount = 0  #กำหนดตัวเก็บค่า
    for char in seq:
        if char == character: 
            charcount = charcount + 1
    return charcount         
```

```python
countChar(DNAseq.iloc[1][0], 'A')   #แถวนี้มีAทั้งหมดกี่ตัว #DNAseq.iloc[1][0]คือ char  #A คือ character
```

```python
countChar('ATTAAAGGTTTATACCTTCCCAGGTAACAAACCAACCAACTTTCGATCTCTTGTAGATCTGTTCTCTAAA', 'A')
```

```python
ANum = 0
TNum = 0
GNum = 0
CNum = 0
for row in range(DNAseq.shape[0]) : #วนตามจำนวนครั้ง
    ANum += countChar(DNAseq.iloc[row][0], 'A')
    TNum += countChar(DNAseq.iloc[row][0], 'T') 
    GNum += countChar(DNAseq.iloc[row][0], 'G') 
    CNum += countChar(DNAseq.iloc[row][0], 'C') 
print(f' There are {ANum} As, {TNum} Ts, {GNum} Gs, {CNum} Cs')    
```

#### ข้อ3 what is the most frequent 5-char-long subsequence? (ลำดับที่ซ้ำกันมากที่สุด5ตัวในDNAนี้คืออะไร)

#### 3.1หาก่อนว่าตัวที่ติดกัน5ตัวมีอะไรบ้าง 3.2 แล้วหาว่าในห้าตัวนั้นมีซ้ำกันกี่ตัว

```python
def populate5charstring(seq) :
    str_list = list()
    for i in range(len(seq)-4) :
         str_list.append(seq[i:i+5]) #เพิ่มด้วย ก็คือตัด ถ้าครบ5 ก็คือมีคอม่ามาคั่น Dna sep(0:5) = (0,1,2,3,4)
    return str_list
```

```python
populate5charstring('ATTAAAGGTTT')
```

```python
populate5charstring(DNAseq.iloc[2][0])  #อ่านแถวที่2คอลัมที่1 ตัด5ตัว
```

```python
list_all_5char = list() #สร้างลิชว่าง
for row in range(DNAseq.shape[0]): #ให้วนลูปให้เท่ากับจำนวนที่0
      list_all_5char = list_all_5char + populate5charstring(DNAseq.iloc[row][0])
```

```python
len(list_all_5char)
```

```python
unique_5char = list(set(list_all_5char)) #ดูว่าซ้ำกันกี่ค่า คำสั่งset จะตัดตัวที่ซ้ำกันออก ให้เป็น ยูนีค
```

```python
len(unique_5char) #ดูว่าในรูปแบบ ซ้ำกันกี่อัน
```

```python
dict_5char = {}   #เป็นดิกเพราะเก็ฐค่าที่เป็นตัวหนังสือเเละตัวเลข
for sub_str in unique_5char:   #ถ้าsub_str = CCTTA    #for1 วน1022
    num = 0
    for s in list_all_5char:   # s = CCTTA    #for2 วน 28000
        if sub_str == s:
            num += 1
    dict_5char[sub_str] = num     #ถ้า sub_str = s 1 ครั้ง num ก็จะเท่ากับ 1    
```

```python
dict_5char
```

```python
maxVal = -100    #คือตัวเลขข้างหลัง  #ตั้งให้น้อยที่สุด ไว้เก็บ
maxKey = None  #คือ จำนวนครั้งข้างหน้า
for k in dict_5char.keys():
    if dict_5char[k] > maxVal :
        maxVal = dict_5char[k]
        maxKey = k
```

```python
maxKey
```

```python
print(f' max 5-long-sequence is {maxKey} eith frequency equals {maxVal}')
```
