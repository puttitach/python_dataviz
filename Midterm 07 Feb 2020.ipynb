#### ข้อ1 สร้าง Function ชื่อ smartDict ซึ่งทำหน้าที่
 -functioon รับ input เป็น list2 ตัวที่มีขนาดเท่ากัน โดย list แรกเป็น list ของตัวเลข list ตัวที่สองเป็น list ของ text
 
 -ถ้าค่าใน list แรก เป็นเลขคี่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่2
 
 -ถ้าค่าใน list แรก เป็นเลขคู่ ให้ค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่1

-output ของ function คือ dict ที่สร้างขึ้น
 
 เช่น output1 = smartdict([1, 2, 3],['apple', 'banana', 'orange']) จะได้ output เป็น {1 :'apple', 'banana' : 2, 3 :'orange'}

output2 = smartDict([7,1,4,6], ['A', 'B', 'E', 'Zzz']) จะได้ {7:'A', 1:'B', 'E':4, 'Zzz :6} 

```python
def smartdict(listnum, listtext) : 
    if listtext : 
        retdict = {}
        for index in range(len(listtext)) :
            retdict[listnum[index]] = listtext[index]
    else :
        retdict = {}
        for index in range(len(listtext)) :
            retdict[listtext[index]] = listnum[index]
    return retdict         
```

```python
smartdict([1, 2, 3],['apple', 'banana', 'orange'])
```

```python
smartDict([7,1,4,6], ['A', 'B', 'E', 'Zzz'])
```

```python
import pandas as pd
```

```python
L = ['ภาษาอีสาน' , 'ภาษาไทย']
Data = pd.DataFrame.from_records([],columns = L)
for i in range(32):
    df = pd.read_excel('data.xlsx', sheet_name = i, usecols = 'B:C', skipfooter = 1) 
    Data = pd.concat([Data,df], axis = 0, ignore_index = True)
```

```python
Data.iloc[0]['ภาษาอีสาน'].replace("/", ",")
```

```python
def replacestr(word):
    co1 = word.replace("/", ",")
    return co1
```

```python
replacestr(Data.iloc[0,1])
```

```python
list_1 = []
list_2 = []
```

```python
for i in range(len(Data)):
    list_1.append(replacestr(Data.iloc[i]['ภาษาอีสาน']))
    list_2.append(replacestr(Data.iloc[i]['ภาษาไทย']))
```

```python
Data['ภาษาอีสานใหม่'] = list_1
Data['ภาษาไทยใหม่'] = list_2
```
