```python
greaterbangkok = df.iloc[4:8,:]  #ตัดตาราง เลือกrowที่ต้องการ กทม สมุทรปราการ นน ปทุม
greaterbangkok
```

โจทย์ : เทียบดูรายได้ของ4จังหวัด ในปี2541ของ

Barchart = กราฟแท่ง #มี input 2ตัว x คือ ตำแหน่งในแนวแกนนอน heigt คือความสูงของกราฟ 

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2])   #คือตำแหน่ง
plt.title('Barchart compareing averange monthly incomes of provinces in Greater Bangkok in 1998')  #เพิ่มชื่อกราฟ  การบ้าน หาวิธีให้โชว์ชื่อกราฟเป็นภาษาไทย
plt.xlabel('provinces')   #เพิ่มชื่อแกนแต่ละแกน แกนx
plt.ylabel('incomes Bath/Month/Househoid')  #เพิ่มชื่อแกนแต่ละแกน แกนy
```

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2], tick_label=greaterbangkok.iloc[:,-1])   #คือตำแหน่ง # tick_label ใส่ชื่อแต่ละแกนโดยเลือกหลักและแถว
plt.title('Barchart compareing averange monthly incomes of provinces in Greater Bangkok in 1998')  #เพิ่มชื่อกราฟ  การบ้าน หาวิธีให้โชว์ชื่อกราฟเป็นภาษาไทย
plt.xlabel('provinces')   #เพิ่มชื่อแกนแต่ละแกน แกนx
plt.ylabel('incomes Bath/Month/Househoid')  #เพิ่มชื่อแกนแต่ละแกน แกนy
```

plot2กราฟ เทียบปี2541กับ2558

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbangkok.iloc[:,-2],width=0.35, tick_label=greaterbangkok.iloc[:,-1], color='m', label='2558' ,align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbangkok.iloc[:,2],width=0.35, tick_label=greaterbangkok.iloc[:,-1], color='y', label='2541' ,align='edge')
plt.title('Barchart compareing averange monthly incomes of provinces in Greater Bangkok in 1998')  #เพิ่มชื่อกราฟ  การบ้าน หาวิธีให้โชว์ชื่อกราฟเป็นภาษาไทย
plt.xlabel('provinces')   #เพิ่มชื่อแกนแต่ละแกน แกนx
plt.legend(loc=0)   #ระบุสีว่าสีนี้ปีไหน แต่ละตำแหน่งมีคำสั่งที่แตกต่างกัน
plt.ylabel('incomes Bath/Month/Househoid')  #เพิ่มชื่อแกนแต่ละแกน แกนy
```

Stacked Barchart Graph กราฟแท่งที่ซ้อนกัน

```python
saledata = pd.DataFrame([{'North':20, 'East':20, 'West':20, 'Sorth':20},    #สมมติข้อมูล
                                            {'North':220, 'East':20, 'West':50, 'Sorth':90}])
saledata
```

บริษัทไหน ขายดีกว่ากัน?

```python
plt.bar([1,2],saledata.iloc[:,0], tick_label=['Company A', 'Company B']) 
plt.bar([1,2],saledata.iloc[:,1], bottom=saledata.iloc[:,0],tick_label=['Company A', 'Company B']) #bottomคือฐานมีอยู่เเล้ว ให้ต่อขึ้นไปซ้อนข้างบน
plt.bar([1,2],saledata.iloc[:,2], bottom=saledata.iloc[:,0]+saledata.iloc[:,1], tick_label=['Company A', 'Company B'])
plt.bar([1,2],saledata.iloc[:,3], bottom=saledata.iloc[:,0]+saledata.iloc[:,1]+saledata.iloc[:,2], tick_label=['Company A', 'Company B'])
```
