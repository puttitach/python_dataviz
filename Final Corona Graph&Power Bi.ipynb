```python
Covrona['Province/State']=Covrona['Province/State'].fillna('Other')
Covrona
```

```python
Covrona_a = Covrona[Covrona['Country']!='Australia']
Covrona_b = Covrona_a[Covrona_a['Country']!='Mainland China']
Covrona_c = Covrona_b[Covrona_b['Country']!='Canada']
Covrona_d = Covrona_c[Covrona_c['Country']!='Mexico']
Covrona_e = Covrona_d[Covrona_d['Country']!='Others']
```

```python
Covrona_New = set( Covrona_d['Country'])
```

```python
Confirmed = []
Deaths = []
Recovered = []
Country = []
for i in Covrona_New:
    Con = 0
    Dea = 0
    Rec = 0
    for j in range(len(Covrona)):
        if i == Covrona['Country'][j]:
            Con+=Covrona['Confirmed'][j]
            Dea+=Covrona['Deaths'][j]
            Rec+=Covrona['Recovered'][j]
    Country.append(i)  
    Confirmed.append(Con) 
    Deaths.append(Dea) 
    Recovered.append(Rec) 
```

```python
Covrona_New1 =  {'Country': Country,'Confirmed':Confirmed,'Deaths':Deaths,'Recovered' :Recovered}
```

```python
Covrona1 = pd.DataFrame.from_dict(Covrona_New1)
Covrona1
```

```python
Covrona1.drop(Covrona1[Covrona1['Country'] == 'Mainland China' ].index, inplace = True)
Covrona1
```
