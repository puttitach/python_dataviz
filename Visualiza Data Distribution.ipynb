```python
import pandas
import numpy as np
import matplotlib
```

```python
irisdata = pandas.read_csv('C:\\Python\\iris.data', header=None)
irisdata
```

```python
setosa_np = irisdata.iloc[ : 50, :-1].to_numpy()
setosa_np[:5,:]
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1])
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=60)
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=setosa_np[ :,2])
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=10*setosa_np[ :,2])
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=setosa_np[ :,2], c = 'c')   #กำหนดสี
```

```python
print (f' max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15 :
        colors.append('b')
    elif i < 0.3 :    
        colors.append('g')
    elif i < 0.45 :
        colors.append('y')
    else :
        colors.append('r')
colors              ###กำหนดค่าให้สรนี้เท่ากับสีอะไร
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = colors)
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = setosa_np[ :,3] )  ###ใช้cเท่ากับแกนที่4
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = setosa_np[ :,3], cmap = 'Set3')
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = np.array([[1,0.5,0]]))    #การผสมสีของrpg
```

เลือกmarker ของจุดplot

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = setosa_np[ :,3], cmap = 'Set3' , marker = "*")
```

visualize 2 Data ลงไปในกราฟเดียวกันเพื่อจะดูว่าสองอันนี้เหมือนและต่างกันยังไง

```python
versicolor_np = irisdata.iloc[ 50 : 100 , :-1].to_numpy()
versicolor_np
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = setosa_np[ :,3], cmap = 'Set3' , marker = "*")
plt.scatter(versicolor_np[ :,0],versicolor_np[ :,1],s=30*versicolor_np[ :,2], c = versicolor_np[ :,3], cmap = 'Set3' , marker = ".")
```

```python
virginica_np = irisdata.iloc[ 50 : 100 , :-1].to_numpy()
virginica_np 
```

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=30*setosa_np[ :,2], c = setosa_np[ :,3], cmap = 'Set3' , marker = "*")
plt.scatter(versicolor_np[ :,0],versicolor_np[ :,1],s=30*versicolor_np[ :,2], c = versicolor_np[ :,3], cmap = 'Set3' , marker = ".")
plt.scatter(virginica_np [ :,0],versicolor_np[ :,1],s=30*virginica_np [ :,2], c = virginica_np [ :,3], cmap = 'Set3' , marker = "|")
```
