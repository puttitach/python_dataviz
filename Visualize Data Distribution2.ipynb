Box plot : plot เพื่อดูการกระจายของ data ตัวที่จำเป็นคือ X -> ลำดับของตัวเลข

```python
import pandas
```

```python
irisdata = pandas.read_csv('C:\\Python\\iris.data', header=None)
irisdata.head()
```

```python
setosa_np = irisdata.iloc[ : 50 , :-1].to_numpy()
versicolor_np = irisdata.iloc[ 50 : 100 , :-1].to_numpy()
virginica_np = irisdata.iloc[ 100 : , :-1].to_numpy()
```

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [15, 5 ]   #ดูการกระจาย
plt.subplot(1, 4, 1)   #มี1 แถว / 4 หลัก / อยู่ตำแหน่งที่ 1
plt.boxplot(setosa_np[:, 0]) 
plt.subplot(1, 4, 2)  
plt.boxplot(setosa_np[:, 1]) 
plt.subplot(1, 4, 3)
plt.boxplot(setosa_np[:, 2]) 
plt.subplot(1, 4, 4)
plt.boxplot(setosa_np[:, 3]) 
plt.show()
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [15, 10 ]  #เอาแต่ละคลาสมาเทียบกัน
plt.boxplot(setosa_np)
```

use boxplot to compare features between 3 classes
ฟีเจอร์แต่ละคลาส 

```python
matplotlib.rcParams[ 'figure.figsize'] = [15, 5 ]   #ดูการกระจาย 
#มี1 แถว / 4 หลัก / อยู่ตำแหน่งที่ 1
plt.subplot(1,4,1)
plt.boxplot(setosa_np[:, 0]) 
plt.boxplot(versicolor_np[:, 0]) 
plt.boxplot(virginica_np[:, 0]) 
plt.subplot(1,4,2)
plt.boxplot(setosa_np[:, 1]) 
plt.boxplot(versicolor_np[:, 1]) 
plt.boxplot(virginica_np[:, 1]) 
plt.subplot(1,4,3)
plt.boxplot(setosa_np[:, 2]) 
plt.boxplot(versicolor_np[:, 2]) 
plt.boxplot(virginica_np[:, 2]) 
plt.subplot(1,4,4)
plt.boxplot(setosa_np[:, 3]) 
plt.boxplot(versicolor_np[:, 3]) 
plt.boxplot(virginica_np[:, 3])
plt.show()
```

อันที่ถูก
ต้องสร้างตัวแปรขึ้นมารับ

```python
matplotlib.rcParams[ 'figure.figsize'] = [15, 5 ]
plt.subplot(1,4,1)
f1 = [setosa_np[:, 0],versicolor_np[:, 0],virginica_np[:, 0]]
plt.boxplot(f1)
plt.subplot(1,4,2)
f2 = [setosa_np[:, 1],versicolor_np[:, 1],virginica_np[:, 1]]
plt.boxplot(f2)
plt.subplot(1,4,3)
f3 = [setosa_np[:, 2],versicolor_np[:, 2],virginica_np[:, 2]]
plt.boxplot(f3)
plt.subplot(1,4,4)
f4 = [setosa_np[:, 3],versicolor_np[:, 3],virginica_np[:, 3]]
plt.boxplot(f4)
plt.show()
```

```python
plt.boxplot(f1)  #ตัวตัวหนังสือที่ไม่สวยงามออก
plt.title('Feature 1 Comparison') ;
```

```python
plt.boxplot(f1,labels=['setosa','versicolor','virginica'])
plt.title('Feature 1 Comparison') ;
```

Bubble Chart ใช้ขนาดวงกลมในการแสดงค่าของdata วงใหญ่=ค่ามาก วงเล็ก=ค่าน้อย

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=1000*setosa_np[ :,2], c = 'y')
```

alpha ค่าโปร่งใส ค่าอยู่ระหว่าง0-1

```python
plt.scatter(setosa_np[ :,0],setosa_np[ :,1],s=1000*setosa_np[ :,2], c = 'y',alpha=0.3)
```

## Map

```python
import pandas 
```

```python
Debt = 'Average Debt.xls.xls'
```

```python
Debt_df = pandas.read_excel(Debt)
```

ขั้นตอนที่2:ทำการดึงข้อมูลจากไฟล์ excel ที่เราต้องการนำมาใช้งาน

```python
Debt_df=pandas.read_excel(Debt,skiprows=3)
Debt_df.head()
```

ขั้นตอนที่3:เลือก column ที่เราสนใจ ทำการตั้งชื่อข้อมูลใหม่เพื่อเอาไว้เก็บข้อมูลที่เราทำการลบออกไปแล้ว เหลือแค่ที่เราต้องการพร้อมสำหรับการใช้งาน

ขั้นตอนที่4:ทำการลบแถวที่ไม่เกี่ยวข้องออกไป

```python
Debt_df[['Region and province', 2549]]
```

```python
Debt_df_debt=Debt_df[['Region and province',2549]]
```

```python
Debt_df_debt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis=0,inplace=True)
```

```python
Debt_df_debt.head()
```

ขั้นตอนที่5: ทำการimport folium เพื่อใช้สำหรับการทำแผนที่

```python
import folium
import pandas as pd
```

```python
data_to_plot = Debt_df_debt
```

ขั้นตอนที่6:เรียกดูหัวcolumn

```python
data_to_plot.head()
```

ขั้นตอนที่7: ทำการDownloadข้อมูลเชิงภูมิศาสตร์ของประเทศไทย "thai json"และทำการลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้

```python
country_geo = "thai.json"
```

```python
new_col=list()
for d in data_to_plot['Region and province'] :
    new_col.append(d.lstrip())
data_to_plot['Region and province'] = new_col    
```

ขั้นตอนที่8: ทำการกำหนดพิกัดที่ต้องการ [10,100]

```python
map = folium.Map(location=[10,100], zoom_start=5)
map
```

ขั้นตอนที่9:ทำการลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือนถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนมีค่ามากจะเป็นสีม่วงเข้มส่วนจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลืองตามแถบรูปข้างบน

```python
map.choropleth(geo_data=country_geo, data=data_to_plot,
              columns=['Region and province', 2549],
              key_on='feature.properties.name',
              fill_color='YlGnBu', fill_opacity=0.7, line_opacity=0.2,
              legend_name= 'Average Debt')
```

```python
map.save('plot_data.html')
```
